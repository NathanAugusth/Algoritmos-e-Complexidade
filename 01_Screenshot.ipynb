```python
import asyncio
from playwright.async_api import async_playwright  # Importa asyncio

async def main():
    async with async_playwright() as p:  # Inicializa o Playwright
        browser = await p.chromium.launch(headless=True)
        page = await browser.new_page()

        await page.goto("https://www.youtube.com")

        await page.wait_for_selector('input[name="search_query"]')

        await page.fill('input[name="search_query"]', "Mrs. Beast")

        await page.press('input[name="search_query"]', "Enter")

        await page.wait_for_selector("ytd-video-renderer", timeout=30000)

        await page.screenshot(path="youtube_results.png", full_page=True)  # Tira um screenshot da página inteira e salva como "youtube_results.png"

        await browser.close()

# Verifica se já existe um loop de eventos rodando
loop = asyncio.get_event_loop()
if loop.is_running():
    task = loop.create_task(main())
    await task
else:
    asyncio.run(main())

from IPython.display import Image
Image("youtube_results.png")  # Exibe a imagem do screenshot no ambiente
```
